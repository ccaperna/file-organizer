```python
import os
import shutil
import csv
import argparse
import sys
import numpy as np
from PIL import Image
from tabulate import tabulate
import ntpath as nt
```

```python
#inizializzo un dizionario e le liste che saranno i valori del diz
diz = dict()
nome = []
tipo = []
grand = []
#memorizzo il percorso sorgente
source = "C:\\FileOrganizer"
fl = os.path.join(source,"files")
os.chdir(source)
#trovo la directory contenente i file di interesse
files = os.listdir(fl)
#creazione cartelle
os.mkdir(os.path.join(fl, "Docs"))
os.mkdir(os.path.join(fl, "Images"))
os.mkdir(os.path.join(fl, "Audio"))
#scorro i file all'interno della dir per processarli 
for file in sorted(files):
    name = os.path.splitext(os.path.basename(file))[0]
    nome.append(name)
    #controllo per creare una nuova dir e spostare il file
    #raccolgo info dai file
    #aggiorno le liste con le informazioni dei file
    if (file.endswith((".txt",".odt"))):
        size = os.path.getsize(os.path.join(fl,file))
        shutil.move(f"{fl}\\{file}", os.path.join(fl, "Docs"))
        print(name, "type: doc", "size: ",size, "B")
        tipo.append("doc")
        grand.append(size)
        print("File spostato con successo")      
    if (file.endswith((".png",".jpeg",".jpg"))):
        size = os.path.getsize(os.path.join(fl,file))
        shutil.move(f"{fl}\\{file}", os.path.join(fl,"Images"))
        print(name, "type: image", "size: ",size, "B")
        tipo.append("image")
        grand.append(size)
        print("File spostato con successo")
    if (file.endswith(".mp3")):
        size = os.path.getsize(os.path.join(fl,file))
        shutil.move(f"{fl}\\{file}", os.path.join(fl,"Audio"))
        print(name, "type: audio", "size: ",size, "B")
        tipo.append("audio")
        grand.append(size)
        print("File spostato con successo")     
#aggiorno il dizionario 
diz["Nome"] = nome
diz["Tipo"] = tipo
diz["Grandezza"] = grand
#creazione documento csv e/o aggiornamento
if os.path.isfile(os.path.join(source,"recap.csv")):
    with open("recap.csv", "w") as recap:
        #creazione oggetto writer per scrivere sul file
        writer = csv.writer(recap)
        #scrivo i valori sulle righe
        writer.writerows(zip(*diz.values()))
else:
    with open("recap.csv", "w") as recap:
        writer = csv.writer(recap)
        #scrivo l'"intestazione" data dalle chiavi del diz
        writer.writerow(diz.keys())
        writer.writerows(zip(*diz.values()))
try:
    #creazione di un oggetto di tipo argparse       
    parser = argparse.ArgumentParser(description = "Sposta un file nella cartella corrispondente")
    #assegnazione argomenti
    parser.add_argument("-f", "--file", type = str, help = "Nome del file + estensione", required = True)
    filename = sys.argv[2]
    args = parser.parse_args()
    #print(type(args))
except argparse.ArgumentError:
    print("Errore, il file non esiste")
    if filename.endswith((".txt",".odt")):
        shutil.move(f"{fl}\\{filename}", os.path.join(fl,"Docs"))
    if filename.endswith(".jpg", ".jpeg", ".png"):
        (f"{fl}\\{filename}", os.path.join(fl,"Images"))
    else:
        (f"{fl}\\{filename}", os.path.join(fl,"Audio"))
```

```python

path = "C:\\FileOrganizer\\files\\Images"
imgs = os.listdir(path)
#dizionario per memorizzare le info delle immagini
diz = dict()
names = []
heigths = []
widths = []
gss = []
rs = []
gs = []
bs = []
alphas = []
for im in imgs:
    name = os.path.splitext(os.path.basename(im))[0]
    names.append(name)
    #caricamento e conversione a np array
    arr = np.asarray(Image.open(os.path.join(path, im)))
    heigths.append(arr.shape[0])
    widths.append(arr.shape[1])
    #grigio = 2dim, rgba = 3dim
    if arr.ndim == 2:
        #aggiorno tutte le liste dei valori del diz
        gss.append(np.mean(arr))
        rs.append(0.00)
        gs.append(0.00)
        bs.append(0.00)
        alphas.append(0.00)
    else:
        gss.append(0.00)
        #aggiungo le colonne che corrispondono ai vari livelli di colore
        rs.append(arr.mean(axis = 0).mean(axis = 0)[0])
        gs.append(arr.mean(axis = 0).mean(axis = 0)[1])
        bs.append(arr.mean(axis = 0).mean(axis = 0)[2])
        if arr.shape[-1] == 3 :
            #non c'Ã¨ il livello alpha
            alphas.append(0.00)        
        else:
            alphas.append(arr.mean(axis = 0).mean(axis = 0)[3])
diz["Name"] = names
diz["Heigth"] = heigths
diz["Width"] = widths
diz["grayscale"] = gss
diz["R"] = rs
diz["G"] = gs
diz["B"] = bs
diz["alpha"] = alphas
#creazione e stampa della tabella
print(tabulate(diz, headers = "keys", floatfmt = ".2f"))
```
